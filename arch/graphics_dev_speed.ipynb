# Why are we slow?

Overall complexity of the system landscape increases exponentially with increasing number of systems, features or dependancies.

```python
# from mpl_toolkits.mplot3d import Axes3D
# from matplotlib.collections import PolyCollection
import matplotlib.pyplot as plt
import numpy as np

SMALL_SIZE = 16
MEDIUM_SIZE = 19
BIGGER_SIZE = 22
FIGURE_WIDTH = 12
FIGURE_HEIGHT = 8

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

plt.xkcd()  # Yes...
```

```python
plt.plot((np.linspace(0, 10)**3))
plt.title('Complexity')
frame1 = plt.gca()
plt.gcf().set_size_inches(FIGURE_WIDTH, FIGURE_HEIGHT)
# frame1.axes.get_xaxis().set_visible(False)
# frame1.axes.get_yaxis().set_visible(False)
frame1.axes.spines['right'].set_visible(False)
frame1.axes.spines['top'].set_visible(False)
frame1.axes.set_xlabel('number of systems/features/dependencies')
frame1.axes.set_ylabel('complexity')
frame1.axes.set_xticklabels([])
frame1.axes.set_yticklabels([])
```

The higher the complexity of the overall system landscape, the lower the development speed since more time is spent for programming itself (more cases / edge cases to take into account) and testing.

```python
plt.plot(1/(np.linspace(0, 10)+2))
plt.title('Development Speed')

frame1 = plt.gca()
plt.gcf().set_size_inches(FIGURE_WIDTH, FIGURE_HEIGHT)
# frame1.axes.get_xaxis().set_visible(False)
# frame1.axes.get_yaxis().set_visible(False)
frame1.axes.spines['right'].set_visible(False)
frame1.axes.spines['top'].set_visible(False)
frame1.axes.set_xlabel('complexity')
frame1.axes.set_ylabel('development speed')
frame1.axes.set_xticklabels([])
frame1.axes.set_yticklabels([])
```

## Operational burden

People believe that work has to be done for development only. However, every new system or dependancy between systems adds something to the overall operational burden. Depending on how well the system is done and integrated into the current system landscape, this surplus can be higher or lower.

```python
plt.plot(np.linspace(0, 10))
plt.plot(np.log(np.linspace(0,10)+1))
plt.hlines(0,0,50)
plt.title('Operational burden')
frame1 = plt.gca()
plt.gcf().set_size_inches(FIGURE_WIDTH, FIGURE_HEIGHT)
# frame1.axes.get_xaxis().set_visible(False)
# frame1.axes.get_yaxis().set_visible(False)
frame1.axes.spines['right'].set_visible(False)
frame1.axes.spines['top'].set_visible(False)
frame1.axes.set_xlabel('number of systems/features/dependencies')
frame1.axes.set_ylabel('operational burden')
frame1.axes.set_ylim(-2,10)
frame1.axes.set_xticklabels([])
frame1.axes.set_yticklabels([])
plt.annotate(
    'status quo',
    xy=(44, 9), arrowprops=dict(arrowstyle='->'), xytext=(25, 9))
plt.annotate(
    'should',
    xy=(45, 2.3), arrowprops=dict(arrowstyle='->'), xytext=(40, 4))
plt.annotate(
    'belief',
    xy=(30, 0), arrowprops=dict(arrowstyle='->'), xytext=(20, -1))
```

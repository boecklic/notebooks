```python
# from mpl_toolkits.mplot3d import Axes3D
# from matplotlib.collections import PolyCollection
from matplotlib import font_manager
font_manager._rebuild()
import matplotlib.pyplot as plt
import numpy as np

plt.xkcd()  # Yes...
plt.plot(1/(np.linspace(0, 10)+2))
plt.title('Development Speed')
frame1 = plt.gca()
# frame1.axes.get_xaxis().set_visible(False)
# frame1.axes.get_yaxis().set_visible(False)
frame1.axes.spines['right'].set_visible(False)
frame1.axes.spines['top'].set_visible(False)
frame1.axes.set_xlabel('complexity')
frame1.axes.set_ylabel('development speed')
frame1.axes.set_xticklabels([])
frame1.axes.set_yticklabels([])
```

```python
plt.plot((np.linspace(0, 10)**3))
plt.title('Complexity')
frame1 = plt.gca()
# frame1.axes.get_xaxis().set_visible(False)
# frame1.axes.get_yaxis().set_visible(False)
frame1.axes.spines['right'].set_visible(False)
frame1.axes.spines['top'].set_visible(False)
frame1.axes.set_xlabel('number of systems/features/dependencies')
frame1.axes.set_ylabel('complexity')
frame1.axes.set_xticklabels([])
frame1.axes.set_yticklabels([])
```

```python
plt.plot(np.linspace(0, 10))
plt.plot(np.log(np.linspace(0,10)+1))
plt.title('Operational burden')
frame1 = plt.gca()
# frame1.axes.get_xaxis().set_visible(False)
# frame1.axes.get_yaxis().set_visible(False)
frame1.axes.spines['right'].set_visible(False)
frame1.axes.spines['top'].set_visible(False)
frame1.axes.set_xlabel('number of systems/features/dependencies')
frame1.axes.set_ylabel('operational burden')
frame1.axes.set_xticklabels([])
frame1.axes.set_yticklabels([])
plt.annotate(
    'status quo',
    xy=(44, 9), arrowprops=dict(arrowstyle='->'), xytext=(25, 9))
plt.annotate(
    'should',
    xy=(45, 2.3), arrowprops=dict(arrowstyle='->'), xytext=(40, 4))
```

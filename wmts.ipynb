```python
import requests

api_url = "https://api3.geo.admin.ch"
service_url = "/rest/services/height"
q_args = "?easting=2624000&elevation_model=COMB&northing=1163000&st=2056"

headers = {}
# headers['referer'] = "vdc.wsl.ch"
headers['referer'] = "map.geo.admin.ch"

print(f"{api_url}{service_url}{q_args}")
resp = requests.get(f"{api_url}{service_url}{q_args}", headers=headers)
print(resp.status_code, resp.text)


```
